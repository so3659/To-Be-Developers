```python
df_jetbrain = pd.read_csv('jetBrain.csv')

grouped_columns = {}
for col in df_jetbrain.columns:
    if '.' in col:
        prefix = col.split('.')[0]
        if prefix not in grouped_columns:
            grouped_columns[prefix] = []
        grouped_columns[prefix].append(col)
    else:
        if 'no_prefix' not in grouped_columns:
            grouped_columns['no_prefix'] = []
        grouped_columns['no_prefix'].append(col)

#     
new_data = pd.DataFrame()

#   
for prefix, cols in grouped_columns.items():
    if prefix != 'no_prefix':
        #      
        sub_df = df_jetbrain[cols]
        # NaN   ,  row    ';' 
        sub_df[prefix] = sub_df.apply(lambda row: ';'.join(row.dropna().astype(str)), axis=1)
        #     
        new_data[prefix] = sub_df[prefix]
    else:
        # .    
        for col in cols:
            new_data[col] = df_jetbrain[col]
```

```python
#        
def split_job_roles(row):
    #    
    keep_patterns = [
        'Developer / Programmer /  Software Engineer',
        'UX / UI Designer',
        'Tester / QA Engineer'
    ]
    
    job_roles = []
    # ;   
    roles = row['job_role'].split(';')
    
    for role in roles:
        #     /  
        if role.strip() not in keep_patterns:
            sub_roles = re.split(r' / ', role)
            for sub_role in sub_roles:
                job_roles.append(sub_role.strip())
        else:
            job_roles.append(role.strip())
    
    new_rows = []
    for job in job_roles:
        new_row = row.copy()
        new_row['job'] = job
        new_rows.append(new_row)
    
    return new_rows
```

```python
junior_spark_df = spark.createDataFrame(junior_expanded_df)
middle_spark_df = spark.createDataFrame(middle_expanded_df)
senior_spark_df = spark.createDataFrame(senior_expanded_df)

def create_specific_job(job_role, target_platforms):
    return (
        when((job_role == 'Developer / Programmer /  Software Engineer') & (target_platforms.contains('Blockchain')), 'BlockChain')
        .when((job_role == 'Developer / Programmer /  Software Engineer') & (target_platforms.contains('Web (Back-end)')) & (target_platforms.contains('Web (Front-end)')), 'Developer, full-stack')
        .when((job_role == 'Developer / Programmer /  Software Engineer') & (target_platforms.contains('Web (Back-end)')), 'Developer, back-end')
        .when((job_role == 'Developer / Programmer /  Software Engineer') & (target_platforms.contains('Web (Front-end)')), 'Developer, front-end')
        .when((job_role == 'Developer / Programmer /  Software Engineer') & (target_platforms.contains('Games')), 'Developer, game or graphics')
        .when((job_role == 'Developer / Programmer /  Software Engineer') & (target_platforms.contains('Mobile')), 'Developer, mobile')
        .when(job_role == 'UX / UI Designer', 'UX / UI Designer')
        .when(job_role == 'Data Engineer', 'Engineer, data')
        .when(job_role == 'DBA', 'Database administrator')
        .when((job_role == 'Tester') | (job_role == 'QA Engineer'), 'Developer, QA or test')
        .when(job_role == 'Data Scientist', 'Data scientist or machine learning specialist')
        .when((job_role == 'Data Analyst') | (job_role == 'Business Analyst'), 'Data or business analyst')
        .when(job_role == 'DevOps Engineer', 'DevOps specialist')
        .otherwise(job_role)
    )

transformed_junibor_df = (
    junior_spark_df.withColumn('job', create_specific_job(col('job'), col('target_platforms')))
                   .withColumn('target_platforms', when(col('job').isin(['BlockChain', 'Developer, full-stack', 'Developer, back-end', 'Developer, front-end', 'Developer, game or graphics', 'Developer, mobile']), col('job')).otherwise(col('target_platforms')))
)

transformed_middle_df = (
    middle_spark_df.withColumn('job', create_specific_job(col('job'), col('target_platforms')))
                   .withColumn('target_platforms', when(col('job').isin(['BlockChain', 'Developer, full-stack', 'Developer, back-end', 'Developer, front-end', 'Developer, game or graphics', 'Developer, mobile']), col('job')).otherwise(col('target_platforms')))
)

transformed_senior_df = (
    senior_spark_df.withColumn('job', create_specific_job(col('job'), col('target_platforms')))
                   .withColumn('target_platforms', when(col('job').isin(['BlockChain', 'Developer, full-stack', 'Developer, back-end', 'Developer, front-end', 'Developer, game or graphics', 'Developer, mobile']), col('job')).otherwise(col('target_platforms')))
)

filtered_job_list = [
    'BlockChain',
    'Developer, full-stack',
    'Developer, back-end',
    'Developer, front-end',
    'Developer, game or graphics',
    'Developer, mobile',
    'UX / UI Designer',
    'Engineer, data',
    'Database administrator',
    'Data scientist or machine learning specialist',
    'Data or business analyst',
    'Product manager',
    'Developer Advocate',
    'Developer, QA or test',
    'DevOps specialist'
]

junior_filtered_by_joblist_df = transformed_junior_df.filter(col('job').isin(filtered_job_list))
middle_filtered_by_joblist_df = transformed_middle_df.filter(col('job').isin(filtered_job_list))
senior_filtered_by_joblist_df = transformed_senior_df.filter(col('job').isin(filtered_job_list))
```
