# 15. 산점도 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시 마이너스 글자 깨짐 현상 방지
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.show()
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s = sizes) # 점 크기
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500  # 1학년 = 500, 2학년 = 1000, 3학년 = 1500
plt.scatter(df['영어'], df['수학'], s = sizes) # 점 크기
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.show()
```

```python
plt.scatter(df['영어'], df['수학'], s = sizes, c = df['학년'], cmap = 'viridis', alpha = 0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.show()
```

```python
plt.figure(figsize=(7, 7))

plt.scatter(df['영어'], df['수학'], s = sizes, c = df['학년'], cmap = 'viridis', alpha = 0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

plt.colorbar(ticks = [1, 2, 3], label = '학년', shrink = 0.5, orientation = 'horizontal')
plt.show()
```
